```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/AAbhishekReddy/covid-trouble/master/data/world.csv')
```

```python
data.shape
```

```python
india = data[data["Country"] == "India"]

india = india[india["New Cases"] > 35000]

# Bar plot
plt.figure(figsize=(30, 20))
plt.bar(x=india["Date"], height=india["New Cases"])
plt.title("India: Number of cases by Date", fontsize=24, fontweight="bold")
plt.savefig("plots/indiacases.png")
```

```python
india = india[['Date', "New Cases"]].copy()
india.head()
```

```python
india.to_csv('india.csv')
```
